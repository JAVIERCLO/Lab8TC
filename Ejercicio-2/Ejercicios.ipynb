Parte b
Implemente y realice profiling con n ∈ {1, 10, 100, 1000, 10000, 100000, 1000000}. Presente
resultados en tabla y gráfica.

```python
import timeit
import matplotlib.pyplot as plt
import pandas as pd
```

```python
"""
    Función del Problema 2 
    Simula el comportamiento del codigo original
"""
def function(n):
    if n <= 1:
        return 0
    
    counter = 0
    for i in range(1, n + 1):  
        for j in range(1, n + 1):  
            counter += 1 
            break 
    
    return counter
```

```python
valores_n = [1, 10, 100, 1000, 10000, 100000, 1000000]

resultados = []

print("Profiling\n")
print("=" * 70)

for n in valores_n:
    # Tiempo de ejecucion
    tiempo_segundos = timeit.timeit(lambda: function(n), number=1)
    tiempo_microsegundos = tiempo_segundos * 1000000  
    
    # Obtener numero de iteraciones
    iteraciones = function(n)
    
    # Guardar resultados
    resultados.append({
        'n': n,
        'Iteraciones': iteraciones,
        'Tiempo (μs)': tiempo_microsegundos,
        'Tiempo (ms)': tiempo_microsegundos / 1000,
        'Tiempo (s)': tiempo_segundos
    })
    
    print(f"n = {n:>7,} | Iteraciones: {iteraciones:>7,} | Tiempo: {tiempo_microsegundos:>10.2f} μs")

print("=" * 70)
print("Profiling completado\n")
```

```python
df = pd.DataFrame(resultados)
print("RESULTADOS")
print("=" * 70)
print(df.to_string(index=False))
print("=" * 70)
```

```python
plt.figure(figsize=(14, 6))

# Gráfica 1: Iteraciones vs n
plt.subplot(1, 2, 1)
plt.plot(df['n'], df['Iteraciones'], marker='o', linewidth=2, markersize=8, color='#4f46e5')
plt.xlabel('n (Tamaño de entrada)', fontsize=12, fontweight='bold')
plt.ylabel('Número de Iteraciones', fontsize=12, fontweight='bold')
plt.title('Iteraciones vs Tamaño de Entrada', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.xscale('log')
plt.yscale('log')

# Gráfica 2: Tiempo vs n
plt.subplot(1, 2, 2)
plt.plot(df['n'], df['Tiempo (μs)'], marker='s', linewidth=2, markersize=8, color='#ef4444')
plt.xlabel('n (Tamaño de entrada)', fontsize=12, fontweight='bold')
plt.ylabel('Tiempo de Ejecución (μs)', fontsize=12, fontweight='bold')
plt.title('Tiempo de Ejecución vs Tamaño de Entrada', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.xscale('log')
plt.yscale('log')

plt.tight_layout()
plt.show()
```
